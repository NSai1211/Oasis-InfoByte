```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
df = pd.read_csv("Unemployment in India.csv")
df.head(5)
```

```python
df.columns
```

```python
df[' Frequency'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['Region',' Date',' Frequency',' Estimated Unemployment Rate (%)',' Estimated Employed',' Estimated Labour Participation Rate (%)','Area'],inplace=True)                        
```

```python
df.isnull().sum()
```

```python
print("row count",df.shape[0])
print("column count",df.shape[1])
```

```python
df[["day", "month", "year"]] = df[' Date'].str.split("-", expand = True)
df.head(5)
```

```python
numeric_columns = df.select_dtypes(include=['number'])
plt.figure(figsize=(8, 6))
sn.heatmap(numeric_columns.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(9,9))
plt.title("Unemployment in india")
sn.histplot(x=' Estimated Unemployment Rate (%)',hue= "Region", data=df)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sn.barplot(x='month',y=' Estimated Unemployment Rate (%)',hue='year',data=df)
plt.show()
```

```python
sn.barplot(x='day',y=' Estimated Unemployment Rate (%)',hue='year',data = df)
```

```python
df[' Date'] = pd.to_datetime(df[' Date'])
plt.figure(figsize=(10, 10))
plt.title("Unemployment in India")
sn.barplot(x=' Date', y=' Estimated Unemployment Rate (%)', hue='Region', data=df)
plt.xticks(rotation=90) 
plt.show()
```
